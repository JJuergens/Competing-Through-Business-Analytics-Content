### Filename: OpiodsVA.csv
Do opioid overdoes tend to be associated with less affluent areasâ€”that is, areas where families have lower incomes?

Some people who start with opioid addictions are reported to transition to heroin use. What is the relationship in Virginia counties between opioid overdoses and heroin overdoses? 

```python
dfOp = pd.read_csv('C:/Users/tetrauja/Desktop/W&M/CTBA/M2/OpidsVA.csv')
```

```python
dfOp.head()
```

```python
fig,ax = plt.subplots()

plt.scatter(dfOp['MedianHouseholdIncome'],dfOp['FPOO-Rate'], alpha = .5, c = 'b')

fig.suptitle('Relationship between Opioid Overdose Rate and Household Income')    
ax.xaxis.set_label_text('Median Household Income')
ax.yaxis.set_label_text('Opiod Overdoses per 100,000 Residents')
fig.set_size_inches(7,5)
plt.show()
```

A scatter plot best represents the relationship between household income and the number of opioid overdoses because they are a number of independent data points and we are looking for trends.

```python
fig,ax = plt.subplots()

plt.scatter(dfOp['FFHO-Rate'],dfOp['FPOO-Rate'], alpha = .5, c = 'g')

fig.suptitle('Relationship between Opioid and Herion Overdose Rates')    
ax.xaxis.set_label_text('Herion Overdoses per 100,000 Residents')
ax.yaxis.set_label_text('Opiod Overdoses per 100,000 Residents')
fig.set_size_inches(7,5)
plt.show()
```

A scatter plot is also the best visual to show the relationship between the opioid and herion overdoes rates because they are a lot of independent data points.

## Filename: presidents.csv
Which states are associated with the greatest number of United States presidents interms of the presidentsâ€™ birthplaces?

```python
dfPres = pd.read_csv('C:/Users/tetrauja/Desktop/W&M/CTBA/M2/presidents.csv')
```

```python
dfPres.head()
```

```python
fig,ax = plt.subplots()
ax.bar(dfPres['State'][:12].unique(),dfPres['State'][:12].value_counts())

fig.suptitle('State Birthplace of Most US Presidents')    
ax.xaxis.set_label_text('# Presidents Birthplace')
ax.yaxis.set_label_text('State')
fig.set_size_inches(10,6)
plt.show()
```

A bar chart is the best visualization to show the number of occurances for something within distinct buckets (i.e. states).
